```python
from basketball_reference_scraper.teams import get_roster, get_team_stats, get_opp_stats, get_roster_stats, get_team_misc
import pandas as pd

# Change TEAM NAME
rs = get_roster_stats('OKC', 2022, data_format='ADVANCED', playoffs=False)

print(rs)

PER = 0.0
for i in range(rs[rs.columns[0]].count()):
  if pd.isna(rs.iloc[i,6]):
    rs.iloc[i,6] = 0.0
  PER = PER + float(rs.iloc[i,5])*float(rs.iloc[i,6])


TS = 0.0
for i in range(rs[rs.columns[0]].count()):
  if pd.isna(rs.iloc[i,7]):
    rs.iloc[i,7] = 0.0
  TS = TS + float(rs.iloc[i,5])*float(rs.iloc[i,7])


TRB = 0.0
for i in range(rs[rs.columns[0]].count()):
  if pd.isna(rs.iloc[i,12]):
    rs.iloc[i,12] = 0.0
  TRB = TRB + float(rs.iloc[i,5])*float(rs.iloc[i,12])


TOV = 0.0
for i in range(rs[rs.columns[0]].count()):
  if pd.isna(rs.iloc[i,16]):
    rs.iloc[i,16] = 0.0
  TOV = TOV + float(rs.iloc[i,5])*float(rs.iloc[i,16])


OWS = 0.0
for i in range(rs[rs.columns[0]].count()):
  if pd.isna(rs.iloc[i,19]):
    rs.iloc[i,19] = 0.0
  OWS = OWS + float(rs.iloc[i,5])*float(rs.iloc[i,19])


DWS = 0.0
for i in range(rs[rs.columns[0]].count()):
  if pd.isna(rs.iloc[i,20]):
    rs.iloc[i,20] = 0.0
  DWS = DWS + float(rs.iloc[i,5])*float(rs.iloc[i,20])


VORP = 0.0
for i in range(rs[rs.columns[0]].count()):
  if pd.isna(rs.iloc[i,27]):
    rs.iloc[i,27] = 0.0
  VORP = VORP + float(rs.iloc[i,5])*float(rs.iloc[i,27])



print('PER:', PER)
print('TS%:', TS)
print('TRB%:', TRB)
print('TOV%:', TOV)
print('OWS:', OWS)
print('DWS:', DWS)
print('VORP:', VORP)
```
